```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("campus_placement_readiness_dirty_dataset.csv")
df.head()
```

```python
df.info()           #Data understanding
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_') #standarize column names
```

```python
df = df.drop_duplicates()        #Remove Duplicates
```

```python
num_cols = ['CGPA', 'Technical_Skills_Score']  #handle missing values: numerical columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df['Internships'] = pd.to_numeric(df['Internships'], errors='coerce')         #Clean internship column
df['Internships'] = df['Internships'].fillna(df['Internships'].median())
```

```python
df['PlacementStatus'] = df['PlacementStatus'].str.strip().str.lower()  #cleaning and standardize target variable

df['PlacementStatus'] = df['PlacementStatus'].replace({
    'yes': 'placed',
    'no': 'notplaced'
})
```

```python
from sklearn.preprocessing import LabelEncoder           #Encode Target Variable

le = LabelEncoder()
df['PlacementStatus_encoded'] = le.fit_transform(df['PlacementStatus'])
```

```python
df.isnull().sum()        #Final Validation
df.duplicated().sum()
```

```python
df.to_csv("campus_placement_cleaned_final.csv", index=False)
```
